```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
median_rent=pd.read_csv("aff_download/ACS_17_5YR_B25071_with_ann.csv",skiprows=[0])
```

```python
gross_rent=pd.read_csv("aff_download/ACS_17_5YR_B25064_with_ann.csv",skiprows=[0])
```

```python
median_rent
```

```python
gross_rent
```

```python
both1=pd.merge(median_rent,gross_rent,on='Id2')
```

```python
both1.columns
```

```python
both2=both1[['Id2', 'Geography_x',
       'Estimate; Median gross rent as a percentage of household income',
       'Estimate; Median gross rent']]
```

```python
both2.columns=both2.columns.str.replace(';','').str.replace(" ",'_').str.lower().str.strip()
```

```python
both2
```

```python
or_map=gpd.read_file('tl_2018_41_tract/tl_2018_41_tract.shp')
```

```python
or_map.columns
```

```python
or_map2=or_map[['GEOID', 'geometry']]
```

```python
or_map2.rename(columns={'GEOID':'id2'},inplace=True)
```

```python
both2.info()
```

```python
or_map2.info()
```

```python
both2['id2']=both2.id2.astype(str)
```

```python
t1=pd.merge(or_map2,both2,on='id2')
```

```python
t1.plot(column='estimate_median_gross_rent_as_a_percentage_of_household_income',cmap='Greens',scheme='quantiles',figsize=(20,20),legend=True)
```
